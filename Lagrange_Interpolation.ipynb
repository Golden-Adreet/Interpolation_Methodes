```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
def leq(x,n,val):
  a = 1
  for i in x:
    if n!=i:
      a*= float((val-i)/(n-i) )
  return a
def lagrange_interpolation(x:list,y:list,val:float):
  sum = 0
  for i in range(len(x)):
    sum+=y[i]*leq(x,x[i],val)
  return sum
```

```python
x=[5,10,12,20,25,30]
y=[45,105,174,259,364,496]
#interpolation using scipy
y2=interp1d(x,y)
#default is linear
y3=interp1d(x,y,kind='cubic')
print (y2)
plt.plot(x,y,'o',x,y2(x),x,y3(x),'-')
val = float(input('Enter a digit to interpolate:'))
print(f'the interpolated value for {val} is:{lagrange_interpolation(x,y,val)}')
plt.show()
```
